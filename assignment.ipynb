```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import pandas as pd

%matplotlib inline
# Always make it pretty.
plt.style.use('ggplot')
font = {'weight': 'bold',
        'size':   16}
plt.rc('font', **font)
```

# Question 1

```python
def flip_coin(n_flips, p=0.5):
    return stats.binom(n=1, p=p).rvs(n_flips)
```

# Question 2

```python
def coin_log_likelihood(p, flips):
    binom = stats.binom(flips, p=p)
    probs = np.array([p, 1-p])
    likelihoods = probs[flips]
    return np.sum(np.log(likelihoods))

#coin_log_likelihood(0.4, flip_coin(10))
```

# Question 3

```python
flip_data = flip_coin(10)
```

# Question 4

```python
log_like25 = coin_log_likelihood(0.25, flip_data)
log_like50 = coin_log_likelihood(0.5, flip_data)
print("log_like25: {}, log_like50: {}".format(log_like25, log_like50))
```

# Question 5

```python
fig, ax = plt.subplots(1, figsize=(10, 10))

ax.bar([1,2],[log_like25, log_like50])
plt.xticks([1,2], ('p=0.25','p=0.5'))
plt.title(flip_data)
```
